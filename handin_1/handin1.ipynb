```python
import pandas as pd


df = pd.read_csv(f'GrossingMovies.csv')

dfDisney = df[df['Distributor'].str.contains("Disney")]
dfDisney.nlargest(n=10, columns=['World Sales (in $)'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

license = df['License'].unique()
distributers = df.iloc[:,10]

PG = []
R = []
G = []

for x in distributers:
    if x == 'PG':
        PG.append(x)
    if x == 'R':
        R.append(x)
    if x == 'G':
        G.append(x)
    if x is None:
        pass


licenseDiff = np.array([len(PG), len(R), len(G)])
ratings = 'PG', 'R', 'G'

plt.pie(licenseDiff, labels=ratings)
plt.show()
```

```python
dfAvg = df.groupby(by='Genre')['World Sales (in $)'].mean()

plt.ticklabel_format(style-'plain')
ax = dfAvg.plot.bar(y='World Sales (in $)')
```
